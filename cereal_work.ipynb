## Does sugar effect calorie count

To investigate this question we should restrict our variables to those related - name, calories and sugar.
After which we can calculate the corrilation between the two and plot that in a scatter for better visability.

```python
s_c_dataframe = clean_cereals_df[["name","calories","sugars"]]
```

- creating new related information DataFrame

```python
import matplotlib.pyplot as plt
```

- importing matplot to allow for graphical representations

```python
correlation = s_c_dataframe['calories'].corr(s_c_dataframe['sugars'])
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(s_c_dataframe['calories'], s_c_dataframe['sugars'], label=f'Correlation: {correlation:.2f}')
plt.title('Correlation between Calories and Sugars')
plt.xlabel('Calories')
plt.ylabel('Sugars')
plt.legend()
plt.grid(True)
plt.legend(loc="upper left")
plt.show()
```

## what we find
> Although this scatter plot does show a positive correlation, it doesn't show a strong one (where being the closer to 1 the stronger the correlation).

>There could be many reasons for this - amounts of fats, types of sugars found or many other reasons.

>However - as this is a fairly positive correlation, we can differ that usually with an increased amount of sugar in a cereal, the calorie count will also be likely to increase (just not with a direct linear relation)

# Does sodium content positively or negatively effect rating

> This is a much better sort of question and will need us to find the correlation between sodium conent and rating. 

- Firstly we will trim the dataframe to the relevant information - name, sodium and rating

```python
s_r_dataframe = cereals_df[["name","sodium","rating"]]
s_r_dataframe.info()
```

- quickly rechecking data for null or missing values

```python
correlation = s_r_dataframe['sodium'].corr(s_r_dataframe['rating'])
```

cacluating the overal correlation between results

```python
plt.figure(figsize=(8, 6))
plt.scatter(s_r_dataframe['sodium'], s_r_dataframe['rating'], label=f'Correlation: {correlation:.2f}')
plt.title('Correlation between Sodium and Rating')
plt.xlabel('Sodium')
plt.ylabel('Rating')
plt.legend()
plt.grid(True)
plt.legend(loc="upper left")
plt.show()
```

> At first glance the negative wide correlation derived from the sample data might indicatte that sodium content doesnt have much effect on the overal rating of the cereal. 

```python
import numpy as np

n = len(s_r_dataframe)
bar_width = 0.35
x_rating = np.arange(n)
x_sodium = x_rating + bar_width

plt.figure(figsize=(12, 6))
plt.bar(x_rating, s_r_dataframe['rating'], width=bar_width, label='Rating')
plt.bar(x_sodium, s_r_dataframe['sodium'], width=bar_width, label='Sodium')
plt.xticks(x_rating + bar_width / 2, s_r_dataframe['name'], rotation=-45, ha='left')

plt.xlabel('Cereal')
plt.ylabel('Value')
plt.title('Cereal Sodium and Ratings')

plt.legend()
plt.show()
```
